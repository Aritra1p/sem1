```python
import pandas as pd
import numpy as np
```

***Rank ordering for Continuous Target***

```python
df=pd.read_excel('Rank ordering data.xlsx',sheet_name="continuous_target")
```

```python
df
```

```python
df['new']=pd.qcut(df['Actual'],10)
df['new'].value_counts()
df['MAPE'] = ((df.Actual-df.Predicted).abs()
                 .div(df.Actual)
                 .cumsum()
                 /np.arange(1,len(df)+1)
             )
df1=df.groupby(['new'])[['Actual','Predicted','MAPE']].mean().sort_values(['new'],ascending=True)
df1
```

***Rank ordering for Binary Target***

```python
df2=pd.read_excel('Rank ordering data.xlsx',sheet_name="binary_target")
```

```python
df2
```

```python
df2['new']=pd.qcut(df2['Predicted'],10)
```

```python
df2['Actual0']=np.where(df2.Actual<1,1,0)
```

```python
df2.rename(columns = {'Actual':'Number of Defaulters','Actual0':'Number of Non-defaulters'}, inplace = True)
```

```python
df3=df2.groupby(['new'])[['Number of Defaulters','Number of Non-defaulters']].sum().sort_values(['new'],ascending=True)
df3
```

```python
df3['Sum']=df3.sum(axis=1)
```

```python
df3["DefRate"] = df3["Number of Defaulters"].div(df3["Sum"].values)
df3
```
